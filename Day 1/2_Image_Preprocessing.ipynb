# Load an image
---
Image pre-processing is best performed using OpenCV and Imutils.


````python
import matplotlib.pyplot as plt
import cv2

original_image = cv2.imread(CourseFolder + '/images/car-224.jpg')
````

---
# Resize image
---
First we resize an image using OpenCV.

````python
import imutils

resized_image = cv2.resize(original_image, (100, 200))

plt.title("Resized image using OpenCV")
plt.imshow(imutils.opencv2matplotlib(resized_image))
plt.show()
````

OpenCV requies the programer to provide the new width and height to be resized to. In certain application, we would like to resize to a certain width/height while maintaining the aspect ratio. 

This resize approach is provided by Imutils.<br>"This ```resize``` function of ```imutils``` maintains the aspect ratio and provides the keyword arguments width and height so the image can be resized to the intended width/height while (1) maintaining aspect ratio and (2) ensuring the dimensions of the image do not have to be explicitly computed by the developer."

````python
import imutils

resized_image_2 = imutils.resize(original_image, height=500)
# or
# resized_image_2 = imutils.resize(original_image, width=100)

plt.title("Resized image using Imutils")
plt.imshow(imutils.opencv2matplotlib(resized_image_2))
plt.show()
````

---
# Image mode conversion
---


````python
HSV_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2HSV)

plt.title("HSV image display as RGB")
plt.imshow(HSV_image)
plt.show()

print("BGR value at (100,150) is ", original_image[100, 150])
print("HSV value at (100,150) is ", HSV_image[100, 150])
````

OpenCV provide a wide range of colour conversion.
* BGR <-> RGB
* RGB <-> HSV
* RGB <-> Gray
* RGB <-> LAB
* RGB <-> YCrCb


---
# Image normalization
---


````python
normalized_image = original_image.astype(float)
normalized_image = normalized_image / 255 

# Covert to to RGB format using 
# Image Slicing method (to be cover in next activity)
normalized_image_RGB = normalized_image[: , :, ::-1] 
plt.title("Normalized image")
plt.imshow(normalized_image_RGB)
plt.show()

print("Original value at (100,150) is ", original_image[100, 150])
print("Normalized value at (100,150) is ", normalized_image[100, 150])
````
